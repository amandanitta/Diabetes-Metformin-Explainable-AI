```python
# drops all identifier columns
health = data.drop(columns=['payer_code','diag_1','diag_2','diag_3',"encounter_id","patient_nbr","admission_type_id","discharge_disposition_id","admission_source_id"])
```

```python
# creates on metformin or not on metformin
health['metformin'] = health['metformin'].replace({0: 0,1 : 1, 2: 1, 3:1})
```

```python
import shap
```

```python
# creates input values and output values with all non-identifying attributes
X = health.drop('readmitted',axis=1)
y = health['readmitted']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

# creates decision tree model
model = DecisionTreeClassifier()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model.fit(X_train,y_train)

# determines test set score
score_test = model.score(X_test,y_test)

print(f'Test Score: {score_test:.4f} ')
```

```python
# determine train set score
score_train = model.score(X_train,y_train)

print(f'Train Score: {score_train:.4f}')
```

```python
# gets explain values
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X)
```

```python
shap.summary_plot(shap_values,X,title="SHAP Summary Ploy")
```

```python
# creates groop of metformin and non-metformin
health_group_re = health.groupby('metformin')
health_group_nometformin = health_group_re.get_group(0)
health_group_metformin = health_group_re.get_group(1)
```

```python
# creates explaination of model for those not on metformin
def drop_info_nonmetformin():
    X = health_group_nometformin.drop('readmitted',axis=1)
    y = health_group_nometformin['readmitted']
    model = DecisionTreeClassifier()
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model.fit(X_train,y_train)
    print(f'Test Accuracy (w/o Metformin): {model.score(X_test,y_test):.4f}')
    print(f'Train accuracy (w/o Metformin) {model.score(X_train,y_train):.4f}')
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X)
    shap.summary_plot(shap_values,X,title="SHAP Summary Ploy")
    
```

```python
drop_info_nonmetformin()
```

```python
# creates explaination of model for those on metformin
def drop_info_metformin():
    X = health_group_metformin.drop('readmitted',axis=1)
    y = health_group_metformin['readmitted']
    model = DecisionTreeClassifier()
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model.fit(X_train,y_train)
    print(f'Test Accuracy (with Metformin): {model.score(X_test,y_test):.4f}')
    print(f'Train accuracy (with Metformin): {model.score(X_train,y_train):.4f}')
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X)
    shap.summary_plot(shap_values,X,title="SHAP Summary Ploy")
```
